# Gaussian Random Number Generator

```python
#import important libraries
import random
import math
import matplotlib.pyplot as plt
```

```python
# take input from user
n = int(input('Enter Number of random want to generator: '))
mu = float(input('mean: '))
st = float(input('standard deviation: '))
a = 1/ (st * math.sqrt(2* math.pi))

# list xaxis, yaxis to store the values, we generate
xaxis=[] # values for xaxis
yaxis=[] # values for yaxis

# n - number of random number to generate

for i in range(n):
    U1 = random.random()
    U2 = random.random()
    x = math.sqrt((-2) * math.log(U1)) * math.cos(2 * math.pi * U2)
    y = a / math.pow(math.e, (pow((x - mu),2) / (2 * pow(st,2))))
    xaxis.append(x)
    yaxis.append(y)
```

```python
# plot the points 

plt.figure(figsize=(20,10))
plt.scatter(xaxis, yaxis)
plt.show()
```
